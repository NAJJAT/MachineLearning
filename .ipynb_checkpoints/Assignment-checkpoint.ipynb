# Define, analyze and prepare: ...

```python
# Load or simulate data
try:
    df = pd.read_csv('mushroom.csv', delimiter=';')
    print(" Dataset loaded from mushroom.csv")
except:
    print("❗ Dataset not found, simulating data...")
    n_samples = 1000
       



# Plot class distribution
sns.countplot(x='class', data=df, palette='viridis')
plt.title("Edible vs Poisonous")
plt.show()

# Plot cap-shape vs class
sns.countplot(x='cap-shape', hue='class', data=df, palette='viridis')
plt.title("Cap Shape vs Edibility")
plt.show()


 # Visualizing missing value counts with more details
missing_values = df.isnull().sum()
missing_values = missing_values[missing_values > 0]
if not missing_values.empty:
    plt.figure(figsize=(12, 6))
    missing_values.plot(kind='bar', color='red', alpha=0.7, edgecolor='k')
    plt.title('Missing Value Counts by Column', fontsize=16)
    plt.xlabel('Columns', fontsize=14)
    plt.ylabel('Missing Count', fontsize=14)
    plt.xticks(rotation=45, fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    for i, value in enumerate(missing_values):
        plt.text(i, value + max(missing_values) * 0.01, str(value), ha='center', fontsize=12)
    plt.show()

# Prepare data
y = df['class']
X = df.drop('class', axis=1)
le_dict = {}
X_encoded = X.copy()
for col in X.columns:
    le = LabelEncoder()
    X_encoded[col] = le.fit_transform(X[col])
    le_dict[col] = le

X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

Mushroom Dataset Analysis

The dataset contains information about mushrooms. Each row describes what a mushroom looks like (such as cap shape, stem color, and other features). The goal is to predict whether a mushroom is edible ('e') or poisonous ('p').

When I looked at the data, I saw that there are slightly more poisonous mushrooms (about 55%) than edible ones (about 45%). This small imbalance can affect how well the model learns, so it's something to keep in mind.

I also checked if some features are linked to edibility. For example, some cap shapes appear more often in poisonous mushrooms. This means that features like cap shape can help the model make better predictions.

One problem I found is that some columns have a lot of missing data. For example, columns like veil-type and spore-print-color are missing more than 90% of their values. These columns are not reliable and will probably need to be removed or ignored during training.

To prepare the data, I separated the target column (class) from the features. I then used LabelEncoder to turn all text values into numbers, since machine learning models need numeric input. Finally, I split the data into a training set (80%) and a test set (20%), using a random seed for reproducibility.

In short: the dataset is useful for training a model because there are clear connections between the features and the target. But I need to carefully handle the columns with lots of missing data, and I will watch out for the slight class imbalance during training and evaluation.

# Model 1: ...

```python
print(" Model 1: Decision Tree")

dt_params = {
    'max_depth': [None, 5, 10],
    'criterion': ['gini', 'entropy']
}
dt_grid = GridSearchCV(DecisionTreeClassifier(random_state=42), dt_params, cv=5)
dt_grid.fit(X_train, y_train)
dt_best = dt_grid.best_estimator_

print("Best parameters:", dt_grid.best_params_)

# Evaluate
y_pred = dt_best.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print(classification_report(y_test, y_pred))

# Extract F1 scores from classification report
report = classification_report(y_test, y_pred, output_dict=True)
f1_score = report['macro avg']['f1-score']  # Using macro average F1

# 1. Confusion Matrix
plt.figure(figsize=(6, 5))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, cmap='Blues', fmt='d')
plt.title("Decision Tree - Confusion Matrix\n(Best params: {})".format(dt_grid.best_params_))
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.show()

# 2. Accuracy & F1 Score Visualization
plt.figure(figsize=(5, 4))
metrics = ['Accuracy', 'F1 Score']
values = [accuracy, f1_score]
colors = ['#66b3ff', '#99ff99']

bars = plt.bar(metrics, values, color=colors, width=0.6)

# Add value labels
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, 
             height - 0.05, 
             f'{height:.4f}', 
             ha='center', 
             va='bottom',
             color='white',
             fontweight='bold')

# Styling
plt.ylim(0, 1.1)
plt.ylabel('Score')
plt.title('Decision Tree - Key Metrics')
plt.grid(axis='y', linestyle='--', alpha=0.3)

# Remove top/right borders
for spine in ['top', 'right']:
    plt.gca().spines[spine].set_visible(False)

plt.tight_layout()
plt.show()
```

Model 1: Decision Tree Output:
I trained a Decision Tree model because it is easy to interpret and performs well on both categorical and numerical data.
After preprocessing (label encoding and splitting the data), I used GridSearchCV to find the best combination of max_depth and criterion.

The model achieved an accuracy of 97%, and the classification report shows that both the "edible" and "poisonous" classes were predicted with high precision and recall.
The confusion matrix shows that very few mushrooms were misclassified which indicates that the model was able to learn clear rules from the features.

This result makes sense because Decision Trees can capture non linear patterns and perform splits based on strong features like odor or spore-print-color, which are often highly predictive in mushroom data.

To avoid overfitting, I limited the depth of the tree to 10 using hyperparameter tuning. This helped the model generalize well on unseen data.

To better understand Decision Trees and how they work, I watched this helpful video:
Decision Tree  Machine Learning Tutorial (YouTube)
The video explains how trees split data, how Gini and Entropy are used, and why pruning or limiting depth is important for avoiding overfitting.

If I wanted to make the model even more robust, I could combine multiple trees using an ensemble method like Random Forest.
But as a first standalone model, the Decision Tree gave reliable and interpretable results.

# Model 2: ...

```python
print(" Model 2: k-Nearest Neighbors")

knn_params = {
    'n_neighbors': [3, 5],
    'weights': ['uniform', 'distance']
}
knn_grid = GridSearchCV(
    KNeighborsClassifier(), 
    knn_params, 
    cv=3,
    verbose=2
)
knn_grid.fit(X_train, y_train)
knn_best = knn_grid.best_estimator_

print("Best parameters:", knn_grid.best_params_)

# Evaluate
y_pred = knn_best.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print(classification_report(y_test, y_pred))

# Extract metrics
report = classification_report(y_test, y_pred, output_dict=True)
f1_score = report['macro avg']['f1-score']

# 1. Confusion Matrix
plt.figure(figsize=(6, 5))
sns.heatmap(confusion_matrix(y_test, y_pred), 
            annot=True, 
            cmap='Purples',  # Different color scheme for KNN
            fmt='d',
            annot_kws={'size': 12})
plt.title(f"k-NN - Confusion Matrix\n(Best params: {knn_grid.best_params_})")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.show()

# 2. Accuracy & F1 Score Visualization
plt.figure(figsize=(5, 4))
metrics = ['Accuracy', 'F1 Score']
values = [accuracy, f1_score]
colors = ['#9d7cbf', '#bc95d9']  # Purple shades for KNN

bars = plt.bar(metrics, values, color=colors, width=0.6)

# Value labels
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, 
             height - 0.05, 
             f'{height:.4f}', 
             ha='center', 
             va='bottom',
             color='white',
             fontweight='bold',
             fontsize=11)

# Styling
plt.ylim(0, 1.1)
plt.ylabel('Score', fontweight='bold')
plt.title('k-NN - Key Metrics', fontweight='bold')
plt.grid(axis='y', linestyle='--', alpha=0.3)

# Clean borders
for spine in ['top', 'right']:
    plt.gca().spines[spine].set_visible(False)

plt.tight_layout()
plt.show()
```

Model 2: k-Nearest Neighbors (k-NN) 
For my second model, I implemented the k-Nearest Neighbors (k-NN) algorithm. I chose this model because it's intuitive and effective for classification tasks, especially when the dataset has clear patterns.

To improve performance, I used GridSearchCV to tune the hyperparameters. The best result was achieved with n_neighbors = 5 and weights = 'distance', which allows the model to give more importance to neighbors that are closer.

The final model reached an accuracy of approximately 82% on the test data. The F1-score was 0.80 for edible mushrooms and 0.84 for poisonous ones, indicating balanced performance across both classes. The confusion matrix confirmed that most predictions were correct.

These results are logical because k-NN performs well when the features are properly preprocessed, as was done here with encoding and scaling. Using distance-based weighting also helped reduce the effect of less relevant neighbors.

# Model 3: ...

```python
print(" Model 3: Linear Support Vector Machine (Fast)")
from sklearn.svm import LinearSVC
from sklearn.model_selection import GridSearchCV

# Hyperparameters for Linear SVM
svm_params = {'C': [0.1, 1, 10]}
svm_grid = GridSearchCV(LinearSVC(dual="auto"), svm_params, cv=3, n_jobs=-1, verbose=1)

print("Training Linear SVM...")
svm_grid.fit(X_train, y_train)
svm_best = svm_grid.best_estimator_

print("Best C:", svm_grid.best_params_)

# Evaluate
y_pred = svm_best.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Extract metrics
report = classification_report(y_test, y_pred, output_dict=True)
f1_score = report['macro avg']['f1-score']

# 1. Confusion Matrix
plt.figure(figsize=(6, 5))
sns.heatmap(confusion_matrix(y_test, y_pred), 
            annot=True, 
            fmt='d', 
            cmap='Oranges',  # Orange theme for SVM
            annot_kws={'size': 12},
            xticklabels=['e', 'p'], 
            yticklabels=['e', 'p'])
plt.title(f"Linear SVM - Confusion Matrix\n(Best C: {svm_grid.best_params_['C']})", pad=20)
plt.xlabel("Predicted", labelpad=10)
plt.ylabel("Actual", labelpad=10)
plt.tight_layout()
plt.show()

# 2. Accuracy & F1 Score Visualization
plt.figure(figsize=(5, 4))
metrics = ['Accuracy', 'F1 Score']
values = [accuracy, f1_score]
colors = ['#ffa64d', '#ffcc99']  # Orange gradient for SVM

bars = plt.bar(metrics, values, color=colors, width=0.6, edgecolor='#ff8000', linewidth=1.2)

# Value labels
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, 
             height - 0.05, 
             f'{height:.4f}', 
             ha='center', 
             va='bottom',
             color='white' if height > 0.7 else '#333333',
             fontweight='bold',
             fontsize=11)

# Styling
plt.ylim(0, 1.1)
plt.ylabel('Score', fontweight='bold')
plt.title('Linear SVM - Key Metrics', fontweight='bold', pad=15)
plt.grid(axis='y', linestyle=':', alpha=0.4)

# Clean borders
for spine in ['top', 'right']:
    plt.gca().spines[spine].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
def plot_learning_curve(estimator, X, y, title):
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=5, train_sizes=np.linspace(0.1, 1.0, 5))

    plt.plot(train_sizes, train_scores.mean(axis=1), 'o-', label="Train")
    plt.plot(train_sizes, test_scores.mean(axis=1), 'o-', label="Validation")
    plt.title(title)
    plt.xlabel("Training Size")
    plt.ylabel("Score")
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot for best model
plot_learning_curve(models[best_idx][1], X_encoded, y, f"Learning Curve - {models[best_idx][0]}")
```
